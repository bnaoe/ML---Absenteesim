## Machine learning modeling 
In this section, we will beging preparing the data for machine learning to help creating a model to predict excessive absenteeism.

```python
#Data Analysis & Machine Learning
data_processed = pd.read_excel('Absenteeism_at_work_processed.xlsx')
data_processed.head()
```

### Create targets
The median absenteeism hours is 3, which will be used to determine which absenses are excessive. Using the median as a cut off line, makes the feature numerically stable and rigid numerically balances our targets.

```python
data_processed['Absenteeism time in hours'].median()
```

```python
#Finds absenteesim hours greater than the median of 3.0 if true return 1 else 0
y_targets = np.where(data_processed['Absenteeism time in hours']>
data_processed['Absenteeism time in hours'].median(),1,0)
y_targets
```

```python
data_processed['Excessive Absence'] = y_targets
```

```python
data_processed.head(1)
```

### Drop columns 
columns that were later determined that has close to zero (0) coffecients in later steps were dropped here. Backward elimination had to be done since coeffecients are not determined until after the model is trained. 

```python
data_with_targets = data_processed.drop(['Absenteeism time in hours','Work load Average/day ','Distance from Residence to Work',
                                        'Hit target','Service time','Disciplinary failure'],axis=1)
data_with_targets.head(1)
```

### Select inputs
Select inputs for regression.

```python
#Do not include the last column Excessive absense as this will be our target
unscaled_inputs =  data_with_targets.iloc[:,0:-1] 
```

### Scale and standardize inputs

```python
#Standardized the data
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy, with_mean, with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns],y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled =X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
#Did not include the dummies to be scaled so not to lose it's unit change of interpretability
columns_to_ommit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']

#added to assign columns to scale by comparing with columns_to_omit using list comprehension
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_ommit]

absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
absenteeism_scaler
```

```python
#This is the scaling mechanism line will calculate the mean and the std dev. of inputs
absenteeism_scaler.fit(unscaled_inputs) 
```

```python
#This operation transforms the unscaled inputs to standardized inputs
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs) 
```

```python
scaled_inputs.head()
```

```python
unscaled_inputs.shape
```

### Train and test split
The data set is split with 80/20 ratio 80% for training the model and 20% for testing the model and the data needs to be shuffled as well.

```python
# Split the data into train & test to train model. We also need to shuffle the data   
from sklearn.model_selection import train_test_split
train_test_split(scaled_inputs,y_targets)
```

```python
#Train_size sets the split usually its 80/20. Random_state 20 make sures the shuffle is the same all the time 
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,y_targets,train_size=0.8,random_state=20)
```

```python
#Verify 80/20 split
print(x_train.shape,y_train.shape) 
print(x_test.shape,y_test.shape)
```

#### Select regression model
I will be using R-squared to determine which model's regression line the data fits closest to. R-squared is a statistical measure of how close the data are to the fitted regression line. It is also known as the coefficient of determination, or the coefficient of multiple determination for multiple regression.
R-squared = Explained variation / Total variation
R-squared is always between 0 and 100%

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor

r2_results = {}

lst_models = [ LinearRegression(),
              SGDRegressor(random_state = 0), 
              KNeighborsRegressor(),    
              GradientBoostingRegressor(random_state = 0),    
              RandomForestRegressor(min_samples_leaf=25, n_estimators=100,
                               warm_start=True, max_features= None,
                               oob_score=True, random_state = 0),
              LogisticRegression(),
              RandomForestClassifier( min_samples_leaf=25, n_estimators=100,
                               warm_start=True, max_features= None,
                               oob_score=True, random_state = 0)]

for lst_model in lst_models:
    
    reg_model = lst_model
    reg_model.fit(x_train, y_train)

    model_name = str(lst_model).split("(")[0]

    r2_results[model_name] = reg_model.score(x_train, y_train)
r2_results
```

```python
# R2 score results
fig, ax = plt.subplots()
ind = range(len(r2_results))
ax.barh(ind, list(r2_results.values()), align='center', 
        color = '#33B7FF', alpha=0.6)
ax.set_title('Comparison of Models')
ax.set_yticks(ind)
ax.set_yticklabels(r2_results.keys())
ax.set_xlabel('R-squared score')
ax.tick_params(left=False, top=False, right=False) 
```

#### Logistic Regression
Training and Testing the Model

I chose to use the logistic regression model as our data fits closest to its regression line. Accuracy of the model will be tested using R-squared and training accuracy is tested with confussion matrix.

```python
#This is the machine learning part
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

reg = LogisticRegression()
reg.fit(x_train,y_train)

# Next few lines are for reviewing accuracy and model significance
#This shows training has an almost 80% accuracy. This means 80% models outputs match the targets.
reg.score(x_train,y_train)
```

```python
model_outputs = reg.predict(x_train)
```

```python
#Since True=1 and False=0 we can find the sum of True values and see how many are True verify reg.score
np.sum((model_outputs==y_train)) /model_outputs.shape[0]
```

#### Check the training model's accuracy
I will use a confusion matrix to determine how many were predicted accurately. Comparing y_train vs predicted outputs. 

```python
# Training Confusion Matrix
import sklearn.metrics as m
cm_df = pd.DataFrame(m.confusion_matrix(model_outputs,y_train))
cm_df.columns = ['Predicted 0','Predicted 1']
cm_df = cm_df.rename(index={0:'Actual 0', 1:'Actual 1'})
cm_df
```

### Get variable coeffecients and the intercept

```python
#get intercept
reg.intercept_
#get coefficients
reg.coef_

feature_name = unscaled_inputs.columns.values
summary_table = pd.DataFrame(columns=['Feature Name'],data = feature_name)
summary_table['Coeffecient'] = np.transpose(reg.coef_)
summary_table.index = summary_table.index + 1
#To add intercept at the top we will shift up all indeces to one
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
#To add intercept at the 0 index
summary_table
```

### Get odds ratio
Coffecients are sorted by the relevance of the problem at hand. If coeffecient is around 0 or odds ratio is around 1, this means corresponding feature is not particularly important. This implies that a coffecient of zero means no matter the feature value it will be zero. For odds, odds increase by a mutiple equal to odds ratio if odds=1 there will be no change as multiplication to one is the same number.

```python
summary_table = summary_table.sort_index()
summary_table['Odds_ratio'] = np.exp(summary_table.Coeffecient)
summary_table.sort_values('Odds_ratio',ascending=False)
```

### Coeffcients and odds ratio analysis

There are four most prominent feature that can determine excessive absenteeism.
Reason_3 which are medical situations needing immediate care such as poisoning shows there's 17% more chances that an excessive absenteeism will occur.
Reason_1 disease related having 15% more chances.
